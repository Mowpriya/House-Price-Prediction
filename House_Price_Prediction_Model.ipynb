## Overview of the Features

* This dataset provides comprehensive information for house price prediction, with 13 column names:

* Price: The price of the house.

* Area: The total area of the house in square feet.

* Bedrooms: The number of bedrooms in the house.

* Bathrooms: The number of bathrooms in the house.

* Stories: The number of stories in the house.

* Mainroad: Whether the house is connected to the main road (Yes/No).

* Guestroom: Whether the house has a guest room (Yes/No).

* Basement: Whether the house has a basement (Yes/No).

* Hot water heating: Whether the house has a hot water heating system (Yes/No).

* Airconditioning: Whether the house has an air conditioning system (Yes/No).

* Parking: The number of parking spaces available within the house.

* Prefarea: Whether the house is located in a preferred area (Yes/No).

* Furnishing status: The furnishing status of the house (Fully Furnished, Semi-Furnished, Unfurnished).

## Separating Independent and dependent features

```python
x=df.drop('price',axis=1)
y=df['price']
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
lr=LinearRegression()
tree=DecisionTreeRegressor(random_state=0)
forest=RandomForestRegressor()
knn=KNeighborsRegressor(n_neighbors=5,metric='minkowski',p=2)
svm=SVR(kernel='linear')
models=[lr,tree,forest,svm,knn]
for model in models:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    mae=mean_absolute_error(y_test,y_pred)
    mse=mean_squared_error(y_test,y_pred)
    r2=r2_score(y_test,y_pred)
    adj_r2=1-(1-r2)*(545-1)/(545-13-1)
    print(type(model).__name__)
    print("******************************************************")
    print("MAE",mae)
    print("---------------------------------------------------")
    print("MSE",mse)
    print("---------------------------------------------------")
    print("R2 score",r2)
    print("----------------------------------------------------")
    print("Adjusted R2",adj_r2)
    print("*******************************************************")
    print("\n")
    
```

## Separating x and y

```python
x=df.drop(['price'],axis=1)
y=df['price']
```

## Train Test split

```python
x_trn,x_tst,y_trn,y_tst=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Feature Scaling

```python
sc=StandardScaler()
x_trn=sc.fit_transform(x_trn)
x_tst=sc.transform(x_tst)
```

```python
x_trn
```

```python
x_tst
```

## Model Building

```python
models = []
models.append(('LinearRegression',LinearRegression()))
models.append(('RandomForest' ,RandomForestRegressor()))
models.append(('Decision Tree ' ,DecisionTreeRegressor()))
models.append (('KNN' ,KNeighborsRegressor()))
models.append(('SVM',SVR()))
```

```python
for name, model in models:
    print(name)
    model.fit(x_trn , y_trn)
    prediction = model.predict(x_tst) 
    print('Training Score:',100*model.score(x_trn,y_trn))
    print('MAE: ' + str(mean_absolute_error(prediction,y_tst)))
    print('_________________________')
    print('MSE:\n ' + str(mean_squared_error(prediction,y_tst)))
    print('_________________________')
    print('R2_score:\n ' + str(r2_score(prediction,y_tst)))
    print('______________________________________')
    print('Adjusted_R2:\n' + str(1-(1-r2_score(prediction,y_tst))*(545-1)/(545-13-1)))
    print('==========================================================================')
```

* After feature importance, we evaluate the model again
* But we do not get better testing result for any algorithms
* So we are performing Hyperparameter tuning to improve the model further
* For Hyperparameter tuning, we are using Grid search Cross valdiation technique

## Grid Search HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

## Hyperparameters for RandomForest, KNN and SVM

```python
params_rf = {'criterion' : ["squared_error", "absolute_error", "friedman_mse", "poisson"],
             'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200]}
```

```python
params_knn= {'n_neighbors': [5,6],'metric':['minkowski','manhattan'],'p':[2,4]}
```

```python
params_svm={'kernel': ('linear', 'rbf','poly'), 'C':[1.5, 10],'gamma': [1e-7, 1e-4],'epsilon':[0.1,0.2,0.5,0.3]}
```

## Hyperparameter tuning for Random Forest Regressor

```python
model_grid_rf = GridSearchCV(RandomForestRegressor(),param_grid=params_rf,verbose = 3).fit(x_trn,y_trn)
model_grid_rf
```

```python
model_grid_rf.score(x_tst,y_tst)
```

* After hyperparameter tuning, random forest regressor testing score is 0.6532709635053864
